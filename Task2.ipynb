```python
 %matplotlib inline
 import numpy as np
 import matplotlib.pyplot as plt
 from random import seed 
 from random import randint
 import time
```

```python

itt = []
ans = []
for i in range(500):
    seed(time.time())


    Xiter = [0]
    itera = 0
    fc = 0
    Ynumone = [1]
    array = np.zeros(1000)
    array[randint(100, 999)] = 1


    while(True):
      itera = itera + 1

      for _ in range(1000):
        if(array[_] == 1):
          j = randint(0, 1)
          array[(_ + j - (1-j) + 1000)%1000] = 1

      ind1 = randint(0, 999)
      ind2 = randint(0, 999)
      temp = array[ind1]
      array[ind1] = array[ind2]
      array[ind2] = temp


      count = 0
      for _ in range(1000):
        count += array[_]

      Xiter.append(itera)
      Ynumone.append(count)

      if count == 1000:
        fc = itera
        break
    ans.append(fc)
    itt.append(i)


plt.plot(itt, ans)
plt.xlabel("iteration")
plt.ylabel("number of number of iterations")
print(np.mean(ans))
# plt.savefig(new.jpg)
```
