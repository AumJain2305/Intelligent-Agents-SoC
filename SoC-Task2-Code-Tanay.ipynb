```
import numpy as np
import matplotlib.pyplot as plt
import random
from random import seed
from random import randint
import time
from scipy.signal import savgol_filter



xaxis = []
x_axis = [0]
yaxis = []
seed(time.time())
nones = [1]
array = np.zeros(10000)
array[4999] = 1
iters = 0

while(True):
  iters += 1
  set1 = np.random.randint(0, 9999, 5)
  set2 = np.random.randint(0, 9999, 5)
  for _ in range(5):
    temp = array[set1[_]]
    array[set1[_]] = array[set2[_]]
    array[set2[_]] = temp
  
  for _ in range(10000):
    if array[_] == 1:
      val1 = random.random()
      val2 = random.random()
      if val1<= 0.35:
        array[(_ + 9999)%10000] = 1
      if val2<= 0.35:
        array[(_ + 1)%10000] = 1
      
  count = 0
  for _ in range(10000):
    count += array[_]
  x_axis.append(iters)
  xaxis.append(iters)
  nones.append(count)
  yaxis.append(count - nones[iters-1])
  if count >= 10000: 
    break


plt.plot(xaxis, yaxis)
plt.xlabel("iteration")
plt.ylabel("dy/dx")
yhat = savgol_filter(yaxis, 51, 2)
plt.plot(xaxis, yhat, color = 'red')
plt.savefig("SoC=Plot3-Tanay.jpg")
maxder = np.amax(yhat)
index = np.where(yhat == np.amax(yhat))
print("max derivative occurs at", index, " with value ", maxder)
```
